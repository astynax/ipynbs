## Игрушечная DataBase

**Цель**: практический опыт использования библиотек **Python** для решения комплексной задачи

**Хочется реализовать:**

- Хранение на диске встроенных структур данных - кортежей и словарей
  - Хранение бинарных данных (например, индексов)
  - Хранение высокоуровневых объектов
- Удобный API для создания/изменения/удаления записей
- Язык описания запросов к базе для выборки/обновления/удаления записей
- Интерфейс командной строки (что-то вроде psql)
- Ссылки между объектами в БД и каскадное извлечение объектов из базы
- Кэширование объектов в ОЗУ
- Конкуррентный доступ и асинхронные операции с БД
- Работа с базой через Socket
- REST-сервис для доступа к БД
