```python
class Zipper(object):
    """
    Класс реализующий интерфейс объекта-"Застёжки",
    позволяющего осуществлять обход некоего набора данных.
    Сам обходчик изображает некую иерархическую структуру,
    позволяющую перемещаться вверх-вниз по уровням
    и влево-вправо в пределах уровня.
    """

    def __init__(self, is_branch, children, make_node, root):
        """
        Параметры:
        @is_branch - предикат, принимающий ноду и
                     возвращающий True, если нода имеет дочерние (нода-ветвь)
        @children  - фукнция, принимающая ноду-ветвь и
                     возвращающая список дочерних нод
        @make_mode - фукнция, принимающая ноду-ветвь и список дочерних нод,
                     возвращающая новую ноду-ветвь, которая заменит переданную
        @root      - контейнер данных, который будет считаться корневой нодой
        """
        self._is_branch = is_branch
        self._children = children
        self._make_node = make_node
        self._node = root
        self._lefts = []
        self._rights = []
        self._path = []
    
    def _fork(self):
        result = self.__class__(
            self._is_branch,
            self._children,
            self._make_node,
            self._node,
        )
        result._path = self._path[:]
        result._lefts = self._lefts[:]
        result._rights = self._rights[:]
        return result

    def __forking(fn):
        def inner(self, *args, **kwargs):
            return fn(self._fork(), *args, **kwargs)
        return inner
    
    @property
    def node(self):
        """
        Возвращает значение текущей ноды
        """
        return self._node

    @property
    def path(self):
        """
        Возвращает список нод - родителей текущей ноды (начиная от "корня")
        """
        return [i[0] for i in self._path]

    @__forking
    def down(self):
        """
        Возвращет новый обходчик, сдвинутый от текущего на один уровень вниз
        (т.е. указывающий на самую левую под-ноду).
        Если двигаться вниз невозможно, возбуждается исключение.
        """
        if not self.is_branch:
            raise TypeError("You can't go down from leaf node!")
        self._path.append((self._node, self._lefts, self._rights))
        children = self._children(self._node)
        if not children:
            raise TypeError("Branch is empty!")
        self._lefts = []
        self._rights = children[::-1]
        self._node = self._rights.pop()
        return self
    
    @__forking
    def up(self):
        """
        Возвращет новый обходчик, сдвинутый от текущего на один уровень вверх
        (т.е. указывающий на родительскую для текщей ноду).
        Если двигаться вверх невозможно, возбуждается исключение.
        """
        if self.is_root:
            raise TypeError("You can't go up from root!")
        children = self._lefts + [self._node] + self._rights[::-1]
        node, self._lefts, self._rights = self._path.pop()
        self._node = self._make_node(node, children)
        return self
    
    @__forking
    def root(self):
        """
        Возвращет новый обходчик, указывающий на корневую ноду.
        """
        while not self.is_root:
            self = self.up()
        return self
    
    @property
    def is_leftmost(self):
        """
        Возвращает True, если текущая нода - крайняя слева среди своих соседей
        """
        return not self._lefts
    
    @property
    def is_rightmost(self):
        """
        Возвращает True, если текущая нода - крайняя справа среди своих соседей
        """
        return not self._rights
    
    @property
    def is_root(self):
        """
        Возвращает True, если текущая нода - корневая
        """
        return not self._path
    
    @property
    def is_branch(self):
        """
        Возвращает True, если текущая нода - ветвь
        """
        return self._is_branch(self._node)
    
    @__forking
    def left(self):
        """
        Возвращет новый обходчик, сдвинутый от текущего влево
        (т.е. указывающий на соседнюю слева ноду).
        Если двигаться влево невозможно, возбуждается исключение.
        """
        if self.is_leftmost:
            raise TypeError("You can't go left from leftmost node!")
        self = self._fork()
        self._rights.append(self._node)
        self._node = self._lefts.pop()
        return self
    
    @__forking
    def right(self):
        """
        Возвращет новый обходчик, сдвинутый от текущего вправо
        (т.е. указывающий на соседнюю справа ноду).
        Если двигаться вправо невозможно, возбуждается исключение.
        """
        if self.is_rightmost:
            raise TypeError("You can't go right from rightmost node!")
        self._lefts.append(self._node)
        self._node = self._rights.pop()
        return self

    @__forking
    def leftmost(self):
        """
        Возвращет новый обходчик, указывающий на крайнюю слева ноду
        """
        # while not self.is_leftmost:
        #     self = self.left()
        if not self.is_leftmost:
            self._rights.append(self._node)
            self._rights.extend(self._lefts[:0:-1])
            self._node = self._lefts[0]
            self._lefts = []
        return self
    
    @__forking
    def rightmost(self):
        """
        Возвращет новый обходчик, указывающий на крайнюю справа ноду
        """
        # while not self.is_rightmost:
        #     self = self.right()
        if not self.is_rightmost:
            self._lefts.append(self._node)
            self._lefts.extend(self._rights[:0:-1])
            self._node = self._rights[0]
            self._rights = []
        return self
    
    @__forking
    def append_child(self, node):
        """
        Возвращает новый обходчик, к дочерним нодам текущей ноды которого
        будет добавлена СПРАВА указанная нода @node.
        Если текущая нода не является ветвью, возбуждается исключение.
        """
        if not self._is_branch(self._node):
            raise TypeError("You can't append child to the leaf node!")
        self._node = self._make_node(
            self._node,
            self._children(self._node) + [node]
        )
        return self

    @__forking
    def insert_child(self, node):
        """
        Возвращает новый обходчик, к дочерним нодам текущей ноды которого
        будет добавлена СЛЕВА указанная нода @node.
        Если текущая нода не является ветвью, возбуждается исключение.
        """
        if not self._is_branch(self._node):
            raise TypeError("You can't append child to the leaf node!")
        self._node = self._make_node(
            self._node,
            [node] + self._children(self._node)
        )
        return self

    @__forking
    def insert_left(self, node):
        """
        Возвращает новый обходчик, слева от текущей ноды которого
        будет вставлена нода @node.
        """
        if self.is_root:
            raise TypeError("Root node can't have the neighbours!")
        self._lefts.append(node)
        return self

    @__forking
    def insert_right(self, node):
        """
        Возвращает новый обходчик, справа от текущей ноды которого
        будет вставлена нода @node.
        """
        if self.is_root:
            raise TypeError("Root node can't have the neighbours!")
        self._rights.append(node)
        return self

    @__forking
    def replace(self, node):
        """
        Возвращает новый обходчик, содержимое текущей ноды которого
        будет заменено на @node.
        """
        self._node = fn(self._node)
        return self
    
    @property
    def is_end(self):
        """
        Возвращает True, если текущая нода не имеет "следующей" (next_node)
        """
        if self.is_root:
            return not self.is_branch # отработка вырожденного случая, когда корень не ветвится
        else:
            if self.is_rightmost:
                if self.is_branch:
                    return False
                else:
                    while True:
                        self = self.up()
                        if self.is_root:
                            return True
                        if not self.is_rightmost:
                            return False
            return False
    
    def next_node(self):
        """
        Возвращает новый обходчик, текущая нода которого
        будет "следующей" по отношению к текущей ноде даного обходчика.
        """
        if self.is_branch:
            return self.down()
        else:
            if self.is_end:
                return self
            elif self.is_rightmost:
                while self.is_rightmost:
                    self = self.up()
            return self.right()

    def __repr__(self):
        return "Zipper[node: %r, path: %r]" % (self.node, self.path)
    
    def __eq__(self, other):
        return (
            self._node == other._node
            and
            self._lefts == other._lefts
            and
            self._rights == other._rights
            and
            self._path == other._path
        )
```
